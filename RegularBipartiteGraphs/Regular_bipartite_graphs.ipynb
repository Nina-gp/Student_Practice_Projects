# Number of Regular bipartite graphs

### What is a regular bipartite graph?
A regular bipartite graph is a type of graph that consists of two disjoint sets of vertices, referred to as "A" and "B," such that every edge in the graph connects a vertex from the A set to a vertex in the B set. Additionally, each vertex in the A set has the same degree, and each vertex in the B set also has the same degree.

#### It is good to be reminded:
Lemma: "If G is a bipartite graph with the bipartition sets X and Y, then the sum of the degrees of the vertices in set X is equal to the sum of the degrees of the vertices in set Y."

According to the above lemma, the number of states of a regular bipartite graph (m, n) with 10 nodes:
- The number of regular bipartite graph (1, 9)= there is just one graph:

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes to the first partition (9 nodes)
nodes_part1 = range(1, 10)
G.add_nodes_from(nodes_part1, bipartite=0)

# Add a node to the second partition (1 node)
G.add_node(10, bipartite=1)

# Add edges between partitions
edges = [(node_part1, 10) for node_part1 in nodes_part1]
G.add_edges_from(edges)

# Display the graph
pos = nx.bipartite_layout(G, nodes=nodes_part1)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=800, font_size=8, font_color='black', font_weight='bold', font_family='sans-serif')
plt.title("regular bipartite graph (1, 9)")
plt.show()
```

- The number of regular bipartite graph (5,5) = there are five such this graph:

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_bipartite_graph(x_degrees, y_degrees):
    # Create a bipartite graph
    G = nx.Graph()

    # Add nodes to each partition
    for i in range(5):
        G.add_node(i, bipartite=0)  # Partition 0 (Left side)
        G.add_node(i + 5, bipartite=1)  # Partition 1 (Right side)

    # Add edges according to specified degrees
    for i in range(5):
        for j in range(x_degrees):
            G.add_edge(i, (i + j + 1) % 5 + 5)

    for i in range(5, 10):
        for j in range(y_degrees):
            G.add_edge(i, (i + j + 1) % 5)

    # Draw the bipartite graph
    pos = {node: (0, i) if node < 5 else (1, i - 5) for i, node in enumerate(G.nodes)}
    nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue')
    plt.show()

# Specify the desired degrees
x_degree = 0  # Number of edges per node in the first partition
y_degree = 1  # Number of edges per node in the second partition

# Draw the graph with the specified degrees
draw_bipartite_graph(x_degree, y_degree)
```
